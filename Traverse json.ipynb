```python
import json
import pandas as pd

# Load the JSON file
json_path = "Sales & Returns Sample v201912.json"
with open(json_path, "r", encoding="utf-8-sig") as f:
    data = json.load(f)

events = data["events"]

# Build a lookup dictionary for fast id -> node access
id_map = {event['id']: event for event in events if 'id' in event}

def find_visual_id(id_map, start_id):
    current_id = start_id
    while current_id:
        node = id_map.get(current_id)
        if not node:
            break
        metrics = node.get('metrics', {})
        visual_id = metrics.get('visualId')
        if visual_id:
            return visual_id
        current_id = node.get('parentId')
    return None

# Find all nodes with name == 'Execute DAX Query'
dax_nodes = [event for event in events if event.get('name') == 'Execute DAX Query']

# For each, find the visualId by traversing up the tree
for node in dax_nodes:
    node_id = node['id']
    visual_id = find_visual_id(id_map, node_id)
    print(f"Node ID: {node_id} -> VisualId: {visual_id}")
```
